```r
# Load required libraries
library(ggplot2)
library(tidyr)

# Set random seed for reproducibility
set.seed(42)

# Create a random dataset of 600 rows and 15 columns between [-100, 100]
random_data <- matrix(runif(600 * 15, min = -100, max = 100), nrow = 600)

# Convert the matrix to a data frame
df <- as.data.frame(random_data)

# (i) Plot scatter graph between Column 5 and Column 6.
ggplot(df, aes(x = V5, y = V6)) +
  geom_point(alpha = 0.6) +
  labs(title = "Scatter Graph: Column 5 vs. Column 6", x = "Column 5", y = "Column 6") +
  theme_minimal()

# (ii) Plot histogram of each column in a single graph.
long_df <- df %>%
  pivot_longer(everything(), names_to = "variable", values_to = "value")

ggplot(long_df, aes(x = value)) +
  geom_histogram(binwidth = 5, alpha = 0.7) +
  facet_wrap(~ variable, scales = "free") +
  labs(title = "Histogram of Each Column", x = "Value", y = "Frequency") +
  theme_minimal()

# (iii) Plot the Box plot of each column in a single graph.
long_df <- df %>%
  pivot_longer(everything(), names_to = "variable", values_to = "value")

ggplot(long_df, aes(x = variable, y = value)) +
  geom_boxplot(alpha = 0.7) +
  labs(title = "Box Plot of Each Column", x = "Column", y = "Value") +
  theme_minimal()

# (iv) Normalize all the values in the dataset between 0 and 10.
normalize <- function(x) {
  (x - min(x)) / (max(x) - min(x)) * 10
}

# Convert the matrix to a data frame and then perform normalization
df <- as.data.frame(random_data)
df_normalized <- df %>%
  mutate(across(everything(), normalize))

# Print the normalized data frame
print(df_normalized)
```

Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```r
# Load required libraries
library(ggplot2)

# Set random seed for reproducibility
set.seed(42)

# Create a random dataset of 500 rows and 5 columns between [5, 10]
random_data <- matrix(runif(500 * 5, min = 5, max = 10), nrow = 500)

# Convert the matrix to a data frame
df <- as.data.frame(random_data)

# (i) Perform t-Test on each column.
t_test_results <- lapply(df, function(col) t.test(col))

# (ii) Perform Wilcoxon Signed Rank Test on each column.
wilcoxon_test_results <- lapply(df, function(col) wilcox.test(col, mu = 7.5))

# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4.

# Select column 3 and column 4 from the data frame
column_3 <- df[, 3]
column_4 <- df[, 4]

# Perform Two Sample t-Test
t_test_3_4 <- t.test(column_3, column_4)

# Perform Wilcoxon Rank Sum Test
wilcoxon_test_3_4 <- wilcox.test(column_3, column_4)

# Print the results of all tests
print("t-Test Results:")
print(t_test_results)

print("\nWilcoxon Signed Rank Test Results:")
print(wilcoxon_test_results)

print("\nTwo Sample t-Test Results for Column 3 and Column 4:")
print(t_test_3_4)

print("\nWilcoxon Rank Sum Test Results for Column 3 and Column 4:")
print(wilcoxon_test_3_4)
```

reate a file that contains 1000 lines of random strings.

```r
# Function to generate a random string of a given length
random_string <- function(length) {
  char_set <- c(letters, LETTERS)
  paste(sample(char_set, length, replace = TRUE), collapse = "")
}

# Number of lines in the file
num_lines <- 1000

# Length of each random string
string_length <- 10

# File name
file_name <- "random_strings.txt"

# Generate random strings
random_strings <- replicate(num_lines, random_string(string_length))

# Write the random strings to the file
writeLines(random_strings, con = file_name)

cat("File '", file_name, "' created with", num_lines, "lines of random strings.\n")
```
