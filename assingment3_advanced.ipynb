Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200]. Perform
the following operations:
(i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number
rows having missing values.
(ii) Replace all the NA values with the average of the column value.
(iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns
having correlation <=0.7.
(iv) Normalize all the values in the dataset between 0 and 10.
(v) Replace all the values in the dataset with 1 if value <=0.5 else with 0

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a random dataset of 100 rows and 30 columns between [1, 200]
np.random.seed(42)
dataset = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)), columns=[f"Col_{i+1}" for i in range(30)])

# (i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of rows having missing values.
dataset.loc[:, "Col_10":"Col_60"] = np.nan
rows_with_missing_values = dataset.isnull().any(axis=1).sum()
print("(i) Count of rows having missing values:", rows_with_missing_values)

# (ii) Replace all the NA values with the average of the column value.
dataset.fillna(dataset.mean(), inplace=True)

# (iii) Find the Pearson correlation among all the columns and plot the heat map. Also select those columns having correlation <= 0.7.
correlation_matrix = dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Selecting columns with correlation <= 0.7
selected_columns = correlation_matrix.columns[correlation_matrix.abs().iloc[:-1, :-1].max() <= 0.7]
print("(iii) Columns with correlation <= 0.7:", selected_columns.tolist())

# (iv) Normalize all the values in the dataset between 0 and 10.
min_value = dataset.min().min()
max_value = dataset.max().max()
dataset_normalized = 10 * (dataset - min_value) / (max_value - min_value)

# (v) Replace all the values in the dataset with 1 if value <= 0.5 else with 0
dataset_binary = dataset_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)

# Printing the modified datasets
print("\nNormalized Dataset between 0 and 10:")
print(dataset_normalized.head())

print("\nBinary Dataset (0 or 1):")
print(dataset_binary.head())
```

Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clusterin

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.spatial.distance import cdist, pdist

# Create a random dataset of 500 rows and 15 columns between [-100, 100]
np.random.seed(42)
dataset = pd.DataFrame(np.random.randint(-100, 101, size=(500, 15)), columns=[f"Col_{i+1}" for i in range(15)])

# (i) Plot scatter graph between Column 5 and Column 6.
plt.figure(figsize=(8, 6))
plt.scatter(dataset["Col_5"], dataset["Col_6"], s=40, alpha=0.6, color='blue')
plt.xlabel("Column 5")
plt.ylabel("Column 6")
plt.title("Scatter Graph: Column 5 vs. Column 6")
plt.grid(True)
plt.show()

# (ii) Plot histogram of each column in a single graph.
plt.figure(figsize=(12, 8))
for col in dataset.columns:
    plt.hist(dataset[col], bins=20, alpha=0.7, label=col)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Each Column")
plt.legend()
plt.grid(True)
plt.show()

# (iii) Plot the Box plot of each column in a single graph.
plt.figure(figsize=(12, 8))
dataset.boxplot(column=dataset.columns.tolist(), vert=False)  # Convert columns to a list
plt.xlabel("Value")
plt.title("Box Plot of Each Column")
plt.grid(True)
plt.show()

# (iv) Perform Hierarchical clustering and plot the distance metric graph.
linkage_methods = ["ward", "complete", "average", "single"]
plt.figure(figsize=(12, 8))

for method in linkage_methods:
    # Calculate pairwise distances instead of distance matrix
    pairwise_distances = pdist(dataset, metric='euclidean')
    clustering = AgglomerativeClustering(n_clusters=None, linkage=method, distance_threshold=0)
    clustering.fit(pairwise_distances)
    distances = cdist(dataset, dataset, metric='euclidean')
    distances = distances[clustering.children_]
    distances = distances.mean(axis=1)
    distances = np.sort(distances, axis=0)
    plt.plot(distances, marker='o', label=method)

plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering: Distance Metric Graph for Different Linkage Methods')
plt.legend()
plt.grid(True)
plt.show()
```

Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column

```python
import numpy as np
import pandas as pd
from scipy import stats

# Create a random dataset of 500 rows and 5 columns between [5, 10]
np.random.seed(42)
dataset = pd.DataFrame(np.random.uniform(5, 10, size=(500, 5)), columns=[f"Col_{i+1}" for i in range(5)])

# (i) Perform t-Test on each column.
t_test_results = {}
for col in dataset.columns:
    t_statistic, p_value = stats.ttest_1samp(dataset[col], popmean=7.5)  # popmean is the expected population mean (7.5 in this case)
    t_test_results[col] = {'t_statistic': t_statistic, 'p_value': p_value}

# (ii) Perform Wilcoxon Signed Rank Test on each column.
wilcoxon_results = {}
for col in dataset.columns:
    statistic, p_value = stats.wilcoxon(dataset[col] - 7.5)  # 7.5 is the hypothesized value
    wilcoxon_results[col] = {'statistic': statistic, 'p_value': p_value}

# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4.
column_3 = dataset['Col_3']
column_4 = dataset['Col_4']

# Two Sample t-Test
t_statistic_2samp, p_value_2samp = stats.ttest_ind(column_3, column_4)

# Wilcoxon Rank Sum Test
statistic_ranksum, p_value_ranksum = stats.ranksums(column_3, column_4)

# Print the results
print("(i) t-Test results for each column:")
for col, result in t_test_results.items():
    print(f"Column {col}: t_statistic={result['t_statistic']:.4f}, p_value={result['p_value']:.4f}")

print("\n(ii) Wilcoxon Signed Rank Test results for each column:")
for col, result in wilcoxon_results.items():
    print(f"Column {col}: statistic={result['statistic']:.4f}, p_value={result['p_value']:.4f}")

print("\n(iii) Two Sample t-Test and Wilcoxon Rank Sum Test between Column 3 and Column 4:")
print(f"Two Sample t-Test: t_statistic={t_statistic_2samp:.4f}, p_value={p_value_2samp:.4f}")
print(f"Wilcoxon Rank Sum Test: statistic={statistic_ranksum:.4f}, p_value={p_value_ranksum:.4f}")
```
